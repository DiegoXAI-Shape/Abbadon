```python
import torch
from utils.models.daowa_maadV3Prueba import Daowa_maadPrueba
from utils.inference.predict import prediccionPrueba

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Cargar modelo
modelo = Daowa_maadPrueba(num_clases=3)

# Cargar pesos viejos directamente, ignorando las quejas de nombres duplicados
state_dict_viejo = torch.load('../Models/ModeloPrueba2026-03-21.pth', map_location=device)
modelo.load_state_dict(state_dict_viejo, strict=False)

modelo.to(device)
modelo.eval()
print("¡Pesos del 10 de Marzo cargados exitosamente al modelo nuevo!")

# Probar con una imagen
prediccionPrueba(modelo, 'data/petimages/cat', '10.jpg', device)
```
